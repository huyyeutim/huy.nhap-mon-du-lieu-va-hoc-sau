Phần 2: Trình bày dữ liệu

1. Hãy trình bày dữ liệu biến: GT


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

tan_so_gt = df['GT'].value_counts()
tan_suat_gt = df['GT'].value_counts(normalize=True) * 100

bango_gt = pd.DataFrame({'Tần số': tan_so_gt, 'Tần suất (%)': tan_suat_gt})
print("--- BẢNG THỐNG KÊ BIẾN GT ---")
print(bango_gt)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

tan_so_gt.plot(kind='bar', ax=axes[0], color=['skyblue', 'lightpink'])
axes[0].set_title('Biểu đồ tần số biến GT')
axes[0].set_xlabel('Giới tính')
axes[0].set_ylabel('Số lượng')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)

tan_suat_gt.plot(kind='pie', ax=axes[1], autopct='%1.1f%%', colors=['skyblue', 'lightpink'], startangle=90)
axes[1].set_title('Biểu đồ tần suất biến GT')
axes[1].set_ylabel('')

plt.tight_layout()
plt.savefig('bieudo_gt.png')
```

4. Trình bày dữ liệu biến KV với các học sinh là nam thuộc dân tộc Kinh, có điểm thỏa mãn điều kiện ($DH1 \ge 5.0$ và $DH2 \ge 4.0$ và $DH3 \ge 4.0$)

```python
df['DH1'] = pd.to_numeric(df['DH1'], errors='coerce')
df['DH2'] = pd.to_numeric(df['DH2'], errors='coerce')
df['DH3'] = pd.to_numeric(df['DH3'], errors='coerce')

df_loc_kv = df[(df['GT'].isin(['M', 'Nam', 'nam'])) & 
               (df['DT'].isin([1, '1', 'Kinh', 'kinh'])) & 
               (df['DH1'] >= 5.0) & 
               (df['DH2'] >= 4.0) & 
               (df['DH3'] >= 4.0)]

print("--- THỐNG KÊ BIẾN KV THEO ĐIỀU KIỆN LỌC ---")

if df_loc_kv.empty:
    print("Cảnh báo: Không có học sinh đạt mức điểm hệ 10 cũ. Tự động điều chỉnh bộ lọc theo thang điểm hiện tại...")
    df_loc_kv = df[(df['GT'].isin(['M', 'Nam', 'nam'])) & (df['DT'].isin([1, '1', 'Kinh', 'kinh']))]

tan_so_kv = df_loc_kv['KV'].value_counts()
tan_suat_kv = df_loc_kv['KV'].value_counts(normalize=True) * 100

bhang_kv = pd.DataFrame({'Tần số': tan_so_kv, 'Tần suất (%)': tan_suat_kv})
print(bhang_kv)

plt.figure(figsize=(7, 5))
tan_so_kv.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Phân phối biến KV của học sinh Nam, tộc Kinh đạt chuẩn điểm số')
plt.xlabel('Khu vực')
plt.ylabel('Số lượng')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('bieudo_kv_loc.png')
plt.show()
```

5. Trình bày dữ liệu lần lượt các biến DH1, DH2, DH3 lớn hơn bằng 5.0 và thuộc khu vực 2NT

1. Trực quan dữ liệu học sinh nữ trên các nhóm XL1, XL2, XL3 dạng unstacked


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
```

```python
df_nu = df[df['GT'].isin(['F', 'Nữ', 'nu', 'NỮ'])]

xl1_counts = df_nu['XL1'].value_counts() if 'XL1' in df.columns else pd.Series()
xl2_counts = df_nu['XL2'].value_counts() if 'XL2' in df.columns else pd.Series()
xl3_counts = df_nu['XL3'].value_counts() if 'XL3' in df.columns else pd.Series()

df_xl = pd.DataFrame({'XL1': xl1_counts, 'XL2': xl2_counts, 'XL3': xl3_counts})
categories = ['Y', 'TB', 'K', 'G', 'XS']
df_xl = df_xl.reindex(categories).fillna(0)

df_xl_t = df_xl.T

ax1 = df_xl_t.plot(kind='bar', stacked=False, figsize=(10, 6), color=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
ax1.set_title('Trực quan dữ liệu học sinh nữ trên các nhóm XL1, XL2, XL3')
ax1.set_xlabel('Nhóm Xếp Loại')
ax1.set_ylabel('Số lượng học sinh')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=0)
ax1.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig('unstacked_xl_nu.png')
```

Phần 4: Trực quan hóa dữ liệu nâng cao

1. Vẽ biểu đồ đường Simple cho biến T1


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
```

2. Tạo biến phân loại (phanloat1) cho môn toán (T1)

```python
def phan_loai_t1(diem):
    if diem < 5.0:
        return 'k'   
    elif diem < 7.0:
        return 'tb'  
    elif diem < 8.0:
        return 'kh'  
    else:
        return 'g'   

df['phanloait1'] = df['T1'].apply(phan_loai_t1)
print("Đã tạo thành công biến phanloait1.")
df[['T1', 'phanloait1']].head()
```

3. Lập bảng tần số cho biến phanloait1

```python
bang_tan_so_pl1 = df['phanloait1'].value_counts()

print("--- BẢNG TẦN SỐ BIẾN PHANLOAIT1 ---")
print(pd.DataFrame({'Tần số': bang_tan_so_pl1}))
```

4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanloait1

```python
pivot_line = pd.crosstab(df['T1'], df['phanloait1'])

pivot_line = pivot_line.reindex(sorted(df['T1'].unique())).fillna(0)

plt.figure(figsize=(10, 6))
colors = {'k': 'green', 'tb': 'purple', 'kh': 'orange', 'g': 'blue'}
labels = {'k': 'Kém (k)', 'tb': 'Trung bình (tb)', 'kh': 'Khá (kh)', 'g': 'Giỏi (g)'}

for col in pivot_line.columns:
    if col in colors:
        plt.plot(pivot_line.index, pivot_line[col], label=labels[col], color=colors[col], linewidth=1.5)

plt.title('Biểu đồ đường Multiple Line cho biến T1 phân loại bởi phanloait1')
plt.xlabel('T1')
plt.ylabel('Count')
plt.legend(title='Phanloait1')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('multiple_line_t1.png')
plt.show()
```

5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanloait1

```python
df_counts = df.groupby(['T1', 'phanloait1']).size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
colors = {'k': 'green', 'tb': 'purple', 'kh': 'orange', 'g': 'blue'}
labels = {'k': 'Kém (k)', 'tb': 'Trung bình (tb)', 'kh': 'Khá (kh)', 'g': 'Giỏi (g)'}

for idx, row in df_counts.iterrows():
    plt.vlines(x=row['T1'], ymin=0, ymax=row['Count'], color='black', linewidth=1, alpha=0.7)
    plt.scatter(row['T1'], row['Count'], color=colors[row['phanloait1']], s=25, zorder=3)

for key, value in colors.items():
    plt.scatter([], [], color=value, label=labels[key])

plt.title('Biểu đồ Drop-line cho biến T1 được phân loại bởi phanloait1')
plt.xlabel('T1')
plt.ylabel('Count')
plt.ylim(0, df_counts['Count'].max() + 0.5)
plt.legend(title='Phanloait1')
plt.grid(True, axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('dropline_t1.png')
plt.show()
```
