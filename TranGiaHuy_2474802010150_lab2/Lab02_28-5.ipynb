1. Vẽ biểu đồ đường Simple cho biến T1


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
```

2. Tạo biến phân loại (phanloat1) cho môn toán (T1)

```python
def phan_loai_t1(diem):
    if diem < 5.0:
        return 'k'   
    elif diem < 7.0:
        return 'tb'  
    elif diem < 8.0:
        return 'kh'  
    else:
        return 'g'   

df['phanloait1'] = df['T1'].apply(phan_loai_t1)
print("Đã tạo thành công biến phanloait1.")
df[['T1', 'phanloait1']].head()
```

3. Lập bảng tần số cho biến phanloait1

```python
bang_tan_so_pl1 = df['phanloait1'].value_counts()

print("--- BẢNG TẦN SỐ BIẾN PHANLOAIT1 ---")
print(pd.DataFrame({'Tần số': bang_tan_so_pl1}))
```

4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanloait1

```python
pivot_line = pd.crosstab(df['T1'], df['phanloait1'])

pivot_line = pivot_line.reindex(sorted(df['T1'].unique())).fillna(0)

plt.figure(figsize=(10, 6))
colors = {'k': 'green', 'tb': 'purple', 'kh': 'orange', 'g': 'blue'}
labels = {'k': 'Kém (k)', 'tb': 'Trung bình (tb)', 'kh': 'Khá (kh)', 'g': 'Giỏi (g)'}

for col in pivot_line.columns:
    if col in colors:
        plt.plot(pivot_line.index, pivot_line[col], label=labels[col], color=colors[col], linewidth=1.5)

plt.title('Biểu đồ đường Multiple Line cho biến T1 phân loại bởi phanloait1')
plt.xlabel('T1')
plt.ylabel('Count')
plt.legend(title='Phanloait1')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('multiple_line_t1.png')
plt.show()
```

5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanloait1

```python
df_counts = df.groupby(['T1', 'phanloait1']).size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
colors = {'k': 'green', 'tb': 'purple', 'kh': 'orange', 'g': 'blue'}
labels = {'k': 'Kém (k)', 'tb': 'Trung bình (tb)', 'kh': 'Khá (kh)', 'g': 'Giỏi (g)'}

for idx, row in df_counts.iterrows():
    plt.vlines(x=row['T1'], ymin=0, ymax=row['Count'], color='black', linewidth=1, alpha=0.7)
    plt.scatter(row['T1'], row['Count'], color=colors[row['phanloait1']], s=25, zorder=3)

for key, value in colors.items():
    plt.scatter([], [], color=value, label=labels[key])

plt.title('Biểu đồ Drop-line cho biến T1 được phân loại bởi phanloait1')
plt.xlabel('T1')
plt.ylabel('Count')
plt.ylim(0, df_counts['Count'].max() + 0.5)
plt.legend(title='Phanloait1')
plt.grid(True, axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('dropline_t1.png')
plt.show()
```

Phần 5: Mô tả dữ liệu và khảo sát dạng phân phối
1. Hãy mô tả và khảo sát phân phối cho biến T1
Gợi ý: 
- Mô tả độ tập trung và phân tán của dữ liệu T1
- Vẽ biểu đồ Box-Plot và xác định các 10 đại lượng trong biểu đồ đó
- Mô tả hình dáng lệch của phân phối T1 dựa vào các đại lượng hướng tâm
- Vẽ biểu đồ Histogram biểu thị hình dáng phân phối
- Mô tả các đặc trưng của phân phối, mức độ lệch và mức độ nhọn
- Kiểm chứng phân phối chuẩn QQ-Plot
- Nhận xét và đánh giá về phân phối của T1

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Đọc dữ liệu
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# Thống kê mô tả
print("--- Thống kê mô tả ---")
print(df["T1"].describe())

# Các đại lượng hướng tâm
print("\n--- Các đại lượng hướng tâm ---")
print("Mean =", df["T1"].mean())
print("Median =", df["T1"].median())
print("Mode =")
print(df["T1"].mode())

# Độ phân tán
print("\n--- Độ phân tán ---")
print("Variance =", df["T1"].var())
print("Standard Deviation =", df["T1"].std())
print("Min =", df["T1"].min())
print("Max =", df["T1"].max())

# Tính các đại lượng Boxplot
Q1 = df["T1"].quantile(0.25)
Q2 = df["T1"].quantile(0.50)
Q3 = df["T1"].quantile(0.75)

IQR = Q3 - Q1

Lower = Q1 - 1.5 * IQR
Upper = Q3 + 1.5 * IQR

print("\n--- Thông số Boxplot ---")
print("Min =", df["T1"].min())
print("Q1 =", Q1)
print("Median =", Q2)
print("Q3 =", Q3)
print("Max =", df["T1"].max())
print("IQR =", IQR)
print("Lower Bound =", Lower)
print("Upper Bound =", Upper)

# Outlier
outlier = df[(df["T1"] < Lower) | (df["T1"] > Upper)]
print("Số lượng Outlier =", len(outlier))

plt.figure(figsize=(6, 4))
plt.boxplot(df["T1"], vert=False)  # vert=False giúp biểu đồ nằm ngang tương tự x=df["T1"] của sns
plt.title("Boxplot của T1")
plt.xlabel("Giá trị T1")
plt.show()

# So sánh Mean và Median
print("\n--- So sánh Mean và Median ---")
mean_T1 = df["T1"].mean()
median_T1 = df["T1"].median()

print("Mean =", mean_T1)
print("Median =", median_T1)

if mean_T1 > median_T1:
    print("Phân phối lệch phải")
elif mean_T1 < median_T1:
    print("Phân phối lệch trái")
else:
    print("Phân phối đối xứng")

# Histogram
plt.figure(figsize=(6, 4))
df["T1"].hist(bins=10)
plt.title("Histogram của T1")
plt.xlabel("T1")
plt.ylabel("Tần số")
plt.show()

# Độ lệch và độ nhọn
print("\n--- Chỉ số phân phối ---")
print("Skewness =", df["T1"].skew())
print("Kurtosis =", df["T1"].kurt())

# QQ Plot
plt.figure(figsize=(6, 4))
stats.probplot(df["T1"], dist="norm", plot=plt)
plt.title("QQ-Plot của T1")
plt.show()

print("""
Nhận xét:
- Dựa vào Histogram và QQ-Plot để đánh giá phân phối.
- Nếu các điểm trên QQ-Plot gần đường thẳng thì dữ liệu gần phân phối chuẩn.
- Skewness > 0: lệch phải.
- Skewness < 0: lệch trái.
- Kurtosis lớn: phân phối nhọn.
- Kurtosis nhỏ: phân phối bẹt.
""")
```

- Trực quan hóa biểu đồ Box-plot, histogram và QQ-plot theo phân nhóm là giá trị
của ‘phanlopT1’.

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Đọc dữ liệu
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# Thống kê mô tả
print("--- Thống kê mô tả ---")
print(df["T1"].describe())

# Các đại lượng hướng tâm
print("\n--- Các đại lượng hướng tâm ---")
print("Mean =", df["T1"].mean())
print("Median =", df["T1"].median())
print("Mode =")
print(df["T1"].mode())

# Độ phân tán
print("\n--- Độ phân tán ---")
print("Variance =", df["T1"].var())
print("Standard Deviation =", df["T1"].std())
print("Min =", df["T1"].min())
print("Max =", df["T1"].max())

# Tính các đại lượng Boxplot
Q1 = df["T1"].quantile(0.25)
Q2 = df["T1"].quantile(0.50)
Q3 = df["T1"].quantile(0.75)

IQR = Q3 - Q1

Lower = Q1 - 1.5 * IQR
Upper = Q3 + 1.5 * IQR

print("\n--- Thông số Boxplot ---")
print("Min =", df["T1"].min())
print("Q1 =", Q1)
print("Median =", Q2)
print("Q3 =", Q3)
print("Max =", df["T1"].max())
print("IQR =", IQR)
print("Lower Bound =", Lower)
print("Upper Bound =", Upper)

# Outlier
outlier = df[(df["T1"] < Lower) | (df["T1"] > Upper)]
print("Số lượng Outlier =", len(outlier))

plt.figure(figsize=(6, 4))
plt.boxplot(df["T1"], vert=False)  # vert=False giúp biểu đồ nằm ngang tương tự x=df["T1"] của sns
plt.title("Boxplot của T1")
plt.xlabel("Giá trị T1")
plt.show()

# So sánh Mean và Median
print("\n--- So sánh Mean và Median ---")
mean_T1 = df["T1"].mean()
median_T1 = df["T1"].median()

print("Mean =", mean_T1)
print("Median =", median_T1)

if mean_T1 > median_T1:
    print("Phân phối lệch phải")
elif mean_T1 < median_T1:
    print("Phân phối lệch trái")
else:
    print("Phân phối đối xứng")

# Histogram
plt.figure(figsize=(6, 4))
df["T1"].hist(bins=10)
plt.title("Histogram của T1")
plt.xlabel("T1")
plt.ylabel("Tần số")
plt.show()

# Độ lệch và độ nhọn
print("\n--- Chỉ số phân phối ---")
print("Skewness =", df["T1"].skew())
print("Kurtosis =", df["T1"].kurt())

# QQ Plot
plt.figure(figsize=(6, 4))
stats.probplot(df["T1"], dist="norm", plot=plt)
plt.title("QQ-Plot của T1")
plt.show()

print("""
Nhận xét:
- Dựa vào Histogram và QQ-Plot để đánh giá phân phối.
- Nếu các điểm trên QQ-Plot gần đường thẳng thì dữ liệu gần phân phối chuẩn.
- Skewness > 0: lệch phải.
- Skewness < 0: lệch trái.
- Kurtosis lớn: phân phối nhọn.
- Kurtosis nhỏ: phân phối bẹt.
""")
```

4. Hãy khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

plt.figure(figsize=(8, 6))

# Lấy danh sách các khu vực duy nhất và định nghĩa bảng màu thủ công
khu_vuc_list = df["KV"].unique()
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown']  # Đảm bảo đủ màu cho các khu vực

for idx, kv in enumerate(khu_vuc_list):
    du_lieu_kv = df[df["KV"] == kv]
    # Vẽ từng nhóm khu vực lên cùng một đồ thị
    plt.scatter(
        du_lieu_kv["T1"], 
        du_lieu_kv["DH1"], 
        label=f"Khu vực {kv}", 
        color=colors[idx % len(colors)], 
        alpha=0.7, 
        edgecolors='w', 
        s=50
    )

plt.title("Biểu đồ phân tán (Scatter Plot) giữa DH1 và T1 theo khu vực")
plt.xlabel("T1")
plt.ylabel("DH1")
plt.legend(title="Khu vực (KV)")
plt.grid(True, linestyle='--', alpha=0.5)  # Thêm lưới mờ giúp dễ quan sát tọa độ điểm
plt.show()

print("--- Hệ số tương quan (Correlation) theo từng khu vực ---")
for i in df["KV"].unique():
    du_lieu = df[df["KV"] == i]
    corr = du_lieu["DH1"].corr(du_lieu["T1"])
    print(f"KV = {i}")
    print(f"Correlation = {corr:.4f}")  # Làm tròn 4 chữ số thập phân để hiển thị đẹp hơn
    print()
```

5. Hãy khảo sát tương quan giữa các biến DH1, DH2, DH3
Gợi ý
- Nhận xét ma trận hiệp phương sai hoặc ma trận tương quan
- Vẽ biểu đồ Scatter giữa các biến

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

print("--- Covariance Matrix ---")
cov_matrix = df[["DH1", "DH2", "DH3"]].cov()
print(cov_matrix)
print()

print("--- Correlation Matrix ---")
corr_matrix = df[["DH1", "DH2", "DH3"]].corr()
print(corr_matrix)
print()

cols = ["DH1", "DH2", "DH3"]
n = len(cols)

# Khởi tạo một khung đồ thị lưới n x n (3x3)
fig, axes = plt.subplots(nrows=n, ncols=n, figsize=(10, 10))

for i in range(n):
    for j in range(n):
        ax = axes[i, j]
        if i == j:
            # Đường chéo chính: Vẽ Histogram của biến đó giống như cặp pairplot mặc định
            ax.hist(df[cols[i]].dropna(), bins=10, edgecolor='black', alpha=0.7)
        else:
            # Các ô còn lại: Vẽ Scatter plot tương quan giữa 2 biến chéo
            ax.scatter(df[cols[j]], df[cols[i]], alpha=0.6, edgecolors='w', s=30)
        
        # Cấu hình hiển thị nhãn trục (chỉ hiện ở rìa ngoài cùng để tránh rối mắt)
        if i == n - 1:
            ax.set_xlabel(cols[j])
        if j == 0:
            ax.set_ylabel(cols[i])
            
        # Tối ưu giao diện lưới
        ax.grid(True, linestyle='--', alpha=0.3)

plt.suptitle("Pairplot của DH1, DH2, DH3 (Vẽ bằng Matplotlib)", y=0.92, fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
