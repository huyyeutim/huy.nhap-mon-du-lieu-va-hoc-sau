## 1. Tiến hành hiểu dữ liệu từ chuyên gia

Tập dữ liệu `titanic_disaster.csv` chứa thông tin về các hành khách trên chuyến tàu Titanic định mệnh. Các thuộc tính cốt lõi bao gồm:
- **PassengerId**: Mã định danh duy nhất của mỗi hành khách.
- **Survived**: Trạng thái sống sót (0 = Không sống sót, 1 = Sống sót).
- **Pclass**: Hạng vé hành khách (1 = Hạng nhất, 2 = Hạng nhì, 3 = Hạng ba).
- **Name**: Họ và tên của hành khách.
- **Sex**: Giới tính của hành khách (male / female).
- **Age**: Tuổi của hành khách.
- **SibSp**: Số lượng anh chị em hoặc vợ/chồng cùng đi trên tàu.
- **Parch**: Số lượng cha mẹ hoặc con cái cùng đi trên tàu.
- **Ticket**: Số hiệu vé tàu.
- **Fare**: Giá vé hành khách đã trả.
- **Cabin**: Số phòng/cabin trên tàu.
- **Embarked**: Cảng lên tàu (C = Cherbourg, Q = Queenstown, S = Southampton).

## 2. Thông thường ta thường xử lý các vấn đề sau về dữ liệu
1. Thiếu dòng tiêu đề ở file csv (`Missing header`)
2. Nhiều biến lưu ở một cột (`Multiple variables in one column`)
3. Dữ liệu cột chứa các giá trị đơn vị không nhất quán (`Inconsistent units`)
4. Dữ liệu có một dòng trống (`Empty rows`)
5. Dữ liệu có các dòng trùng lặp (`Duplicate rows`)
6. Các ký tự không phải ASCII (`Non-ASCII characters`)
7. Giá trị bị mất (`Missing values`)
8. Tiêu đề cột là giá trị chứ không phải tên biến (`Column headers are values`)

## 3. Vấn đề 1: Tiến hành tải dữ liệu vào chương trình ứng dụng Python và giải quyết vấn đề “Missing header in the csv file”
*Giải thích:* Tập dữ liệu `titanic_disaster.csv` đã có sẵn dòng tiêu đề chuẩn ở hàng đầu tiên. Khi dùng `pd.read_csv()`, thư viện sẽ tự động nhận diện chính xác cấu trúc các cột mà không làm lệch dữ liệu.

```python
import pandas as pd
import numpy as np

# Tải dữ liệu từ tệp tin csv thực tế
df = pd.read_csv('titanic_disaster.csv')

print("--- Hiển thị cấu trúc 5 dòng đầu tiên của dữ liệu Titanic ---")
display(df.head())
print(f"Kích thước dữ liệu ban đầu: {df.shape}")
```

## 4. Vấn đề 2: Nhiều biến lưu ở một cột
*Giải thích:* Trong cột `Name`, dữ liệu chứa cả Họ (Surname), Danh xưng (Title như Mr, Mrs, Miss) và Tên của hành khách. Chúng ta có thể phân tách thông tin danh xưng để phục vụ cho các phân tích chuyên sâu hơn.

```python
# Trích xuất danh xưng (Title) từ cột Name bằng biểu thức chính quy (Regex)
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

print("--- Các danh xưng được phân tách từ một cột ra ngoài ---")
print(df['Title'].value_counts().head(10))
display(df[['Name', 'Title']].head())
```

## 5. Vấn đề 3: Dữ liệu cột chứa các giá trị đơn vị không nhất quán
*Giải thích:* Đồng bộ hóa dữ liệu dạng chuỗi ký tự định tính như `Sex` và `Embarked` về chữ in hoa chuẩn, loại bỏ khoảng trắng dư thừa để loại trừ rủi ro phân loại sai sót.

```python
# Loại bỏ khoảng trắng và chuẩn hóa chữ in hoa
df['Sex'] = df['Sex'].str.strip().str.upper()
df['Embarked'] = df['Embarked'].str.strip().str.upper()

print("Giá trị phân loại duy nhất cột Sex:", df['Sex'].unique())
print("Giá trị phân loại duy nhất cột Embarked:", df['Embarked'].unique())
```

## 6. Vấn đề 4: Dữ liệu có một dòng trống
*Giải thích:* Tiến hành rà soát các hàng trống hoàn toàn dữ liệu trong bảng (chứa toàn bộ giá trị `NaN`) để xóa bỏ triệt để.

```python
# Kiểm tra số dòng trống hoàn toàn
all_null_rows = df.isnull().all(axis=1).sum()
print(f"Số lượng dòng trống hoàn toàn phát hiện: {all_null_rows}")

# Loại bỏ dòng trống (nếu có)
df.dropna(how='all', inplace=True)
df.reset_index(drop=True, inplace=True)
print(f"Kích thước tập dữ liệu sau xử lý dòng trống: {df.shape}")
```

## 7. Vấn đề 5: Dữ liệu có các dòng trùng lặp
*Giải thích:* Loại bỏ các bản ghi trùng lặp thông tin hoàn chỉnh để bảo toàn tính độc lập của các mẫu dữ liệu.

```python
# Đếm và xóa các dòng trùng lặp
duplicate_rows = df.duplicated().sum()
print(f"Số lượng dòng trùng lặp phát hiện: {duplicate_rows}")

df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
print(f"Kích thước tập dữ liệu sau khi xóa trùng lặp: {df.shape}")
```

## 8. Vấn đề 6: Các ký tự không phải ASCII
*Giải thích:* Loại bỏ hoặc mã hóa lại các ký tự đặc biệt lạ có trong văn bản (ví dụ như các ký tự unicode phức tạp trong tên hành khách) về chuẩn bảng mã ASCII.

```python
def remove_non_ascii(text):
    if pd.isna(text):
        return text
    return str(text).encode("ascii", errors="ignore").decode().strip()

# Áp dụng cho các cột dữ liệu dạng chữ ký tự
for col in ['Name', 'Ticket', 'Cabin']:
    df[col] = df[col].apply(remove_non_ascii)

print("Đã hoàn thành rà soát và chuẩn hóa bảng mã ký tự ASCII.")
```

## 9. Vấn đề 7: Giá trị bị mất
*Giải thích:* Tập dữ liệu Titanic thực tế bị khuyết thiếu ở các cột `Age`, `Cabin`, và `Embarked`. 
- Chúng ta sẽ trực quan hóa dữ liệu thiếu bằng biểu đồ Heatmap.
- Xem xét phân phối Tuổi theo Hạng vé (Pclass) qua biểu đồ Boxplot để đưa ra phương án điền khuyết chính xác theo từng nhóm.

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("--- Số lượng ô thiếu dữ liệu ban đầu ---")
print(df.isnull().sum())

# 1. Trực quan hóa dữ liệu thiếu bằng Heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Biểu đồ nhiệt (Heatmap) thể hiện các vị trí dữ liệu bị khuyết thiếu')
plt.show()

# 2. Trực quan hóa mối quan hệ giữa Pclass và Age bằng Boxplot trước khi điền khuyết
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Age', data=df, palette='Set2')
plt.title('Biểu đồ hộp (Boxplot) phân phối Độ tuổi theo Hạng vé (Pclass)')
plt.xlabel('Hạng vé (Pclass)')
plt.ylabel('Tuổi (Age)')
plt.show()

print("\n-> Nhận xét: Tuổi của hành khách có sự chênh lệch rõ rệt theo từng hạng vé.")
print("Do đó, ta sẽ điền khuyết cột Age theo giá trị Trung vị (Median) của từng Pclass để đảm bảo tính chính xác.\n")

# 3. Tiến hành điền khuyết thông minh theo nhóm Pclass
df['Age'] = df['Age'].fillna(df.groupby('Pclass')['Age'].transform('median'))

# 4. Điền khuyết giá trị cảng lên tàu (Embarked) bằng Yếu vị (Mode)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# 5. Điền khuyết cột Cabin bằng một giá trị đại diện riêng biệt biệt vì tỉ lệ thiếu quá cao
df['Cabin'] = df['Cabin'].fillna('UNKNOWN')

print("--- Kiểm tra lại số lượng ô thiếu dữ liệu sau khi điền khuyết ---")
print(df.isnull().sum())

# Trực quan hóa lại sau khi làm sạch để chứng minh dữ liệu đã hoàn toàn sạch
plt.figure(figsize=(10, 4))
sns.heatmap(df[['Age', 'Embarked', 'Cabin']].isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Heatmap kiểm tra lại sau khi đã xử lý điền khuyết')
plt.show()
```

## 10. Vấn đề 8: Tiêu đề cột là giá trị chứ không phải tên biến
*Giải thích:* Đồng nhất kiểu dữ liệu định lượng cho toàn bộ bảng. Chuyển đổi cột `Age` từ kiểu float64 về kiểu dữ liệu số nguyên `int64` chuẩn xác để dễ dàng xây dựng mô hình mạng nơ-ron học sâu.

```python
# Chuyển đổi thuộc tính Age về kiểu số nguyên nguyên bản
df['Age'] = df['Age'].astype('int64')

print("--- Kiểm tra thông tin cấu trúc dữ liệu hoàn chỉnh cuối cùng ---")
df.info()

print("\n--- Hiển thị 10 dòng dữ liệu Titanic sạch lý tưởng ---")
display(df.head(10))

# Vẽ biểu đồ đếm số lượng hành khách sống sót theo giới tính để làm giàu trực quan cho bài Lab
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Sex', data=df, palette='coolwarm')
plt.title('Thống kê số lượng sống sót (1) và thiệt mạng (0) phân theo Giới tính')
plt.xlabel('Trạng thái sống sót')
plt.ylabel('Số lượng hành khách')
plt.show()

# Xuất tập dữ liệu sạch ra tệp tin mới để phục vụ lưu trữ
df.to_csv('cleaned_titanic_disaster.csv', index=False)
print("\nĐã tạo thành công file dữ liệu sạch: 'cleaned_titanic_disaster.csv'")
```
