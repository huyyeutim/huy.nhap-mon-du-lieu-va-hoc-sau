4. Thống kê dữ liệu thiếu cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: Mặc định thiếu thì điền giá trị 0

```python
print("\nTHỐNG KÊ DỮ LIỆU THIẾU CHO CỘT DÂN TỘC:")
# Xem tên các cột
print(df.columns)

# Thống kê dữ liệu thiếu cột dân tộc
missing = df["DT"].isnull().sum()

print("Số lượng dữ liệu thiếu:", missing)

# Điền giá trị 0
df["DanToc"] = df["DT"].fillna(0)

print(df["DT"].isnull().sum())
```

5. Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử dụng phương pháp Mean

```python
print("\nTHỐNG KÊ DỮ LIỆU CHO BIẾN T1:")
# Đếm số lượng dữ liệu thiếu
missing_T1 = df["T1"].isnull().sum()

print("Số lượng dữ liệu thiếu của T1:", missing_T1)

# Lập bảng tần số
tan_so = df["T1"].isnull().value_counts()

print("\nBảng tần số:")
print(tan_so)

# Lập bảng tần suất
tan_suat = df["T1"].isnull().value_counts(normalize=True) * 100

print("\nBảng tần suất (%):")
print(tan_suat)

# Tính giá trị Mean của cột T1
mean_T1 = df["T1"].mean()

print("\nGiá trị Mean của T1:", mean_T1)

# Thay thế dữ liệu thiếu bằng Mean
df["T1"] = df["T1"].fillna(mean_T1)

print("\nSau khi thay thế dữ liệu thiếu:")

# Kiểm tra lại dữ liệu thiếu
print(df["T1"].isnull().sum())
```

6. Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.

```python

# Chọn các cột điểm số kiểu số
score_columns = [col for col in df.columns 
                 if df[col].dtype in ['int64', 'float64']]

# Thay thế dữ liệu thiếu bằng Mean
for col in score_columns:
    
    # Kiểm tra số lượng thiếu
    missing = df[col].isnull().sum()
    
    if missing > 0:
        mean_value = df[col].mean()
        df[col] = df[col].fillna(mean_value)
        
        print(f"{col}: đã thay {missing} giá trị thiếu bằng Mean = {mean_value}")
```

7. Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12.
• Công thức tính: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10

```python
print("\nTÍNH TRUNG BÌNH MÔN:")
# TBM lớp 10
df["TBM1"] = (df["T1"]*2 + df["L1"] + df["H1"] + df["S1"] +
              df["V1"]*2 + df["X1"] + df["D1"] + df["N1"]) / 10

# TBM lớp 11
df["TBM2"] = (df["T2"]*2 + df["L2"] + df["H2"] + df["S2"] +
              df["V2"]*2 + df["X2"] + df["D2"] + df["N2"]) / 10

# TBM lớp 12
df["TBM3"] = (df["T3"]*2 + df["L3"] + df["H3"] + df["S3"] +
              df["V3"]*2 + df["X3"] + df["D3"] + df["N3"]) / 10


# In kết quả
print(df[["TBM1", "TBM2", "TBM3"]].head())
```

8. Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 10, 11, 12 như sau:
• Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
• Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
• Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
• Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
• Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)

```python
# 8. Tạo biến xếp loại XL1, XL2, XL3

print("\nXẾP LOẠI HỌC LỰC")

# Hàm xếp loại
def xep_loai(tbm):
    if tbm < 5:
        return "Y"
    elif tbm < 6.5:
        return "TB"
    elif tbm < 8:
        return "K"
    elif tbm < 9:
        return "G"
    else:
        return "XS"

# Tạo biến xếp loại
df["XL1"] = df["TBM1"].apply(xep_loai)
df["XL2"] = df["TBM2"].apply(xep_loai)
df["XL3"] = df["TBM3"].apply(xep_loai)

# In kết quả
print(df[["TBM1", "XL1", "TBM2", "XL2", "TBM3", "XL3"]].head())
```

9. Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương pháp Min-Max Normalization

```python

print("\nCHUYỂN ĐỔI THANG ĐIỂM")

# Công thức Min-Max:
# X_new = ((X - Min) / (Max - Min)) * 4

# Tạo biến US_TBM1
df["US_TBM1"] = ((df["TBM1"] - df["TBM1"].min()) /
                 (df["TBM1"].max() - df["TBM1"].min())) * 4

# Tạo biến US_TBM2
df["US_TBM2"] = ((df["TBM2"] - df["TBM2"].min()) /
                 (df["TBM2"].max() - df["TBM2"].min())) * 4

# Tạo biến US_TBM3
df["US_TBM3"] = ((df["TBM3"] - df["TBM3"].min()) /
                 (df["TBM3"].max() - df["TBM3"].min())) * 4


# In kết quả
print(df[["US_TBM1", "US_TBM2", "US_TBM3"]].head())
```

10. Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3 như sau
• Với khối A, A1 nếu [(DH1*2 + DH2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt
• Với khối B nếu [(DH1 + DH2*2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt
• Với khối khác nếu [(DH1+ DH2 + DH3)/3] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt

```python
print("\nXÉT TUYỂN ĐẠI HỌC")

# Hàm xét tuyển
def xet_tuyen(row):

    khoi = row["KT"]

    # Khối A, A1
    if khoi in ["A", "A1"]:
        diem = (row["DH1"]*2 + row["DH2"] + row["DH3"]) / 4

    # Khối B
    elif khoi == "B":
        diem = (row["DH1"] + row["DH2"]*2 + row["DH3"]) / 4

    # Các khối khác
    else:
        diem = (row["DH1"] + row["DH2"] + row["DH3"]) / 3

    # Đậu / Rớt
    if diem >= 5:
        return 1
    else:
        return 0


# Tạo cột KQXT
df["KQXT"] = df.apply(xet_tuyen, axis=1)

# In kết quả
print(df[["KT", "DH1", "DH2", "DH3", "KQXT"]].head())
```
