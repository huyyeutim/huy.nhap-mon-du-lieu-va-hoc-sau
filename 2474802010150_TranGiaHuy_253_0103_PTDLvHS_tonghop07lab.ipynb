# LAB 01: BÀI THỰC HÀNH THAO TÁC DỮ LIỆU
**Nội dung:** Thao tác dữ liệu điểm thi đại học của học sinh
**Dữ liệu:** `dulieuxettuyendaihoc.csv` - điểm trung bình môn, khu vực, khối thi và điểm thi đại học của 100 học sinh


## Phần 01: Xác định và phân loại dữ liệu định tính và định lượng


```python

import pandas as pd
import numpy as np

df = pd.read_csv('dulieuxettuyendaihoc.csv')
print("Kích thước dữ liệu:", df.shape)
print("\nCác cột trong dữ liệu:")
print(list(df.columns))

```

```python

# 1. Xác định cột định lượng (kiểu số) và định tính (kiểu object)
quantitative = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]
qualitative = [col for col in df.columns if col not in quantitative]

print("Cột định lượng (Quantitative):")
print(quantitative)
print(f"\nSố lượng: {len(quantitative)}")
print("\nCột định tính (Qualitative):")
print(qualitative)
print(f"\nSố lượng: {len(qualitative)}")

```

## Phần 02: Định nghĩa các thang đo phù hợp cho từng biến số


```python

print("THANG ĐO CỦA TỪNG BIẾN:")
for col in df.columns:
    if df[col].dtype == 'object':
        print(f"  {col}: Định danh (Nominal)")
    else:
        print(f"  {col}: Tỉ lệ (Ratio)")

```

## Phần 03: Tải dữ liệu và in 10 dòng đầu, 10 dòng cuối


```python

print("10 dòng đầu tiên:")
print(df.head(10))
print("\n" + "="*80)
print("10 dòng cuối cùng:")
print(df.tail(10))

```

## Phần 04: Thống kê dữ liệu thiếu cột DT (dân tộc) và hiệu chỉnh


```python

print("THỐNG KÊ DỮ LIỆU THIẾU CHO CỘT DÂN TỘC:")
missing_DT = df["DT"].isnull().sum()
print(f"Số lượng dữ liệu thiếu cột DT: {missing_DT}")

# Bảng tần số
print("\nBảng tần số dữ liệu thiếu:")
print(df["DT"].isnull().value_counts())

# Bảng tần suất
print("\nBảng tần suất (%):")
print(df["DT"].isnull().value_counts(normalize=True) * 100)

# Số giá trị riêng biệt
print(f"\nSố giá trị riêng biệt (unique) của DT: {df['DT'].nunique()}")

# Thay thế bằng giá trị 0
df["DanToc"] = df["DT"].fillna(0)
print(f"\nSau khi điền 0, còn thiếu: {df['DanToc'].isnull().sum()}")

```

## Phần 05: Thống kê dữ liệu thiếu biến T1 và hiệu chỉnh bằng Mean


```python

print("THỐNG KÊ DỮ LIỆU THIẾU CHO BIẾN T1:")
missing_T1 = df["T1"].isnull().sum()
print(f"Số lượng dữ liệu thiếu của T1: {missing_T1}")

# Bảng tần số
print("\nBảng tần số:")
print(df["T1"].isnull().value_counts())

# Bảng tần suất
print("\nBảng tần suất (%):")
print(df["T1"].isnull().value_counts(normalize=True) * 100)

# Mean
mean_T1 = df["T1"].mean()
print(f"\nGiá trị Mean của T1: {mean_T1:.2f}")

# Thay thế
df["T1"] = df["T1"].fillna(mean_T1)
print(f"Sau khi thay thế, còn thiếu: {df['T1'].isnull().sum()}")

```

## Phần 06: Xử lý tất cả dữ liệu thiếu cho các biến điểm số còn lại


```python

score_columns = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]
print("XỬ LÝ DỮ LIỆU THIẾU BẰNG MEAN:")
for col in score_columns:
    missing = df[col].isnull().sum()
    if missing > 0:
        mean_val = df[col].mean()
        df[col] = df[col].fillna(mean_val)
        print(f"  {col}: đã thay {missing} giá trị thiếu bằng Mean = {mean_val:.2f}")
print("\nHoàn tất xử lý dữ liệu thiếu!")

```

## Phần 07: Tạo biến TBM1, TBM2, TBM3 (trung bình môn)


```python

print("TÍNH TRUNG BÌNH MÔN:")
print("Công thức: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10")

df["TBM1"] = (df["T1"]*2 + df["L1"] + df["H1"] + df["S1"] +
              df["V1"]*2 + df["X1"] + df["D1"] + df["N1"]) / 10
df["TBM2"] = (df["T2"]*2 + df["L2"] + df["H2"] + df["S2"] +
              df["V2"]*2 + df["X2"] + df["D2"] + df["N2"]) / 10
df["TBM3"] = (df["T3"]*2 + df["L3"] + df["H3"] + df["S3"] +
              df["V3"]*2 + df["X3"] + df["D3"] + df["N3"]) / 10

print(df[["TBM1", "TBM2", "TBM3"]].head())

```

## Phần 08: Tạo biến xếp loại XL1, XL2, XL3


```python

def xep_loai(tbm):
    if tbm < 5: return "Y"
    elif tbm < 6.5: return "TB"
    elif tbm < 8: return "K"
    elif tbm < 9: return "G"
    else: return "XS"

df["XL1"] = df["TBM1"].apply(xep_loai)
df["XL2"] = df["TBM2"].apply(xep_loai)
df["XL3"] = df["TBM3"].apply(xep_loai)

print(df[["TBM1", "XL1", "TBM2", "XL2", "TBM3", "XL3"]].head())

```

## Phần 09: Chuyển đổi thang điểm 10 sang thang điểm 4 (Min-Max)


```python

print("CHUYỂN ĐỔI THANG ĐIỂM (MIN-MAX NORMALIZATION)")
print("Công thức: X_new = ((X - Min) / (Max - Min)) * 4")

df["US_TBM1"] = ((df["TBM1"] - df["TBM1"].min()) /
                 (df["TBM1"].max() - df["TBM1"].min())) * 4
df["US_TBM2"] = ((df["TBM2"] - df["TBM2"].min()) /
                 (df["TBM2"].max() - df["TBM2"].min())) * 4
df["US_TBM3"] = ((df["TBM3"] - df["TBM3"].min()) /
                 (df["TBM3"].max() - df["TBM3"].min())) * 4

print(df[["US_TBM1", "US_TBM2", "US_TBM3"]].head())

```

## Phần 10: Tạo biến KQXT (kết quả xét tuyển)


```python

print("TẠO BIẾN KẾT QUẢ XÉT TUYỂN (KQXT)")

def tinh_kqxt(row):
    dh1, dh2, dh3 = row["DH1"], row["DH2"], row["DH3"]
    kt = row["KT"]
    if kt in ["A", "A1"]:
        diem = (dh1*2 + dh2 + dh3) / 4
    elif kt == "B":
        diem = (dh1 + dh2*2 + dh3) / 4
    else:
        diem = (dh1 + dh2 + dh3) / 3
    return 1 if diem >= 5 else 0

df["KQXT"] = df.apply(tinh_kqxt, axis=1)
print(f"Số thí sinh đậu: {df['KQXT'].sum()}/{len(df)}")
print(df[["KT", "DH1", "DH2", "DH3", "KQXT"]].head(10))

```

## Phần 11: Lưu trữ dữ liệu đã xử lý


```python

df.to_csv("processed_dulieuxettuyendaihoc.csv", index=False)
print("Đã lưu file: processed_dulieuxettuyendaihoc.csv")
print("Kích thước:", df.shape)

```

### 02. Tạo biến phân loại phanloait1


```python

def phan_loai_t1(diem):
    if diem < 5.0: return 'k'
    elif diem < 7.0: return 'tb'
    elif diem < 8.0: return 'kh'
    else: return 'g'

df['phanloait1'] = df['T1'].apply(phan_loai_t1)
print("Bảng tần số phanloait1:")
print(df['phanloait1'].value_counts())
print("\nDữ liệu mẫu:")
print(df[['T1', 'phanloait1']].head(10))

```

```python

df_melted['PulseRate'] = pd.to_numeric(df_melted['PulseRate'], errors='coerce')

def hierarchical_fill(row, group_df):
    if not pd.isna(row['PulseRate']):
        return row['PulseRate']
    # 1. Mean của same Id
    indiv = group_df[group_df['Id'] == row['Id']]['PulseRate'].mean()
    if not pd.isna(indiv):
        return indiv
    # 2. Mean theo Sex
    sex_val = group_df[group_df['Sex'] == row['Sex']]['PulseRate'].mean()
    if not pd.isna(sex_val):
        return sex_val
    # 3. Giá trị y khoa mặc định
    return 72.0

df_melted['PulseRate'] = df_melted.apply(lambda r: hierarchical_fill(r, df_melted), axis=1)
print("Đã điền khuyết toàn bộ cột PulseRate.")
print(df_melted.head())

```

# LAB 05: NHẬN DẠNG QUẦN ÁO GIÀY DÉP THỜI TRANG VỚI FASHION-MNIST
**Thư viện:** PyTorch
**Kiến trúc:** Mạng nơ-ron kết nối đầy đủ (Fully Connected)


### 01. Import thư viện và tải dữ liệu


```python

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

transform = transforms.Compose([transforms.ToTensor()])

trainset = torchvision.datasets.FashionMNIST(root='./data', train=True,
                                             download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

testset = torchvision.datasets.FashionMNIST(root='./data', train=False,
                                            download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False)

classes = ['T-shirt/top','Trouser','Pullover','Dress','Coat',
           'Sandal','Shirt','Sneaker','Bag','Ankle boot']

print(f"Train: {len(trainset)} images, Test: {len(testset)} images")

```

### 02. Định nghĩa mô hình


```python

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(28*28, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 10)

    def forward(self, x):
        x = x.view(-1, 28*28)
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

net = Net()
print(net)

```

### 03. Huấn luyện mô hình


```python

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters(), lr=0.001)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
net.to(device)
print(f"Device: {device}")

for epoch in range(5):
    running_loss = 0.0
    for images, labels in trainloader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = net(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {running_loss/len(trainloader):.4f}")

print("Huấn luyện xong!")

```

### 04. Đánh giá mô hình


```python

correct = 0
total = 0
net.eval()
with torch.no_grad():
    for images, labels in testloader:
        images, labels = images.to(device), labels.to(device)
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Độ chính xác trên tập test: {100 * correct / total:.2f}%")

```

### 05. Hiển thị kết quả dự đoán


```python

def imshow(img):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)), cmap='gray')
    plt.axis('off')
    plt.show()

dataiter = iter(testloader)
images, labels = next(dataiter)
images, labels = images.to(device), labels.to(device)
outputs = net(images)
_, predicted = torch.max(outputs, 1)

for i in range(5):
    imshow(images[i].cpu())
    print(f"Thực tế: {classes[labels[i]]} | Dự đoán: {classes[predicted[i]]}")

```

# LAB 06: XÂY DỰNG MẠNG NƠ-RON SÂU (CNN)
**Mục tiêu:** Xây dựng mạng CNN cho bài toán nhận dạng FashionMNIST, so sánh với mạng NN đơn giản ở Lab 5


### 01. Import và tải dữ liệu


```python

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import DataLoader

transform = transforms.Compose([transforms.ToTensor()])

trainset = torchvision.datasets.FashionMNIST(root='./data', train=True,
                                             download=True, transform=transform)
trainloader = DataLoader(trainset, batch_size=64, shuffle=True)

testset = torchvision.datasets.FashionMNIST(root='./data', train=False,
                                            download=True, transform=transform)
testloader = DataLoader(testset, batch_size=64, shuffle=False)

classes = ['T-shirt/top','Trouser','Pullover','Dress','Coat',
           'Sandal','Shirt','Sneaker','Bag','Ankle boot']
print(f"Train: {len(trainset)}, Test: {len(testset)}")

```

### 02. Định nghĩa mạng CNN sâu


```python

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # Block 1: Conv + ReLU + MaxPool
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool1 = nn.MaxPool2d(2, 2)  # 28 -> 14

        # Block 2
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.pool2 = nn.MaxPool2d(2, 2)  # 14 -> 7

        # Block 3
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)
        self.pool3 = nn.MaxPool2d(2, 2)  # 7 -> 3

        # Dropout
        self.dropout = nn.Dropout(0.25)

        # Fully connected
        self.fc1 = nn.Linear(128 * 3 * 3, 256)
        self.fc2 = nn.Linear(256, 10)

    def forward(self, x):
        x = self.pool1(torch.relu(self.bn1(self.conv1(x))))
        x = self.pool2(torch.relu(self.bn2(self.conv2(x))))
        x = self.pool3(torch.relu(self.bn3(self.conv3(x))))
        x = x.view(-1, 128 * 3 * 3)
        x = self.dropout(x)
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

model = CNN()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print(f"Device: {device}")
print(model)
total_params = sum(p.numel() for p in model.parameters())
print(f"Total parameters: {total_params:,}")

```

### 03. Huấn luyện CNN


```python

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

train_losses = []
for epoch in range(5):
    running_loss = 0.0
    model.train()
    for images, labels in trainloader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    epoch_loss = running_loss / len(trainloader)
    train_losses.append(epoch_loss)
    print(f"Epoch {epoch+1}, Loss: {epoch_loss:.4f}")

print("Huấn luyện CNN xong!")

```

### 04. Đánh giá CNN


```python

correct = 0
total = 0
model.eval()
with torch.no_grad():
    for images, labels in testloader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"CNN - Độ chính xác trên tập test: {100 * correct / total:.2f}%")

```

### 05. Hiển thị kết quả CNN

## Phần 08: Phân tích cảm xúc người dùng


```python

from nltk.sentiment import SentimentIntensityAnalyzer

# Sử dụng VADER cho sentiment analysis
sia = SentimentIntensityAnalyzer()

# Một số câu mẫu
test_sentences = [
    "This movie was absolutely fantastic! I loved every minute of it.",
    "The product is terrible, completely waste of money.",
    "It's okay, nothing special but not bad either.",
    "I am so happy with the results, exceeded my expectations!",
    "This is the worst experience I have ever had."
]

print("PHÂN TÍCH CẢM XÚC (SENTIMENT ANALYSIS):")
for sentence in test_sentences:
    scores = sia.polarity_scores(sentence)
    compound = scores['compound']
    if compound >= 0.05:
        sentiment = "Tích cực 😊"
    elif compound <= -0.05:
        sentiment = "Tiêu cực 😞"
    else:
        sentiment = "Trung tính 😐"
    print(f"\nText: {sentence}")
    print(f"Scores: {scores}")
    print(f"Kết luận: {sentiment}")

```

### Naive Bayes với Movie Reviews


```python

import random
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy as nltk_accuracy

# Lấy dữ liệu movie reviews
try:
    reviews = nltk.corpus.movie_reviews
    documents = [(list(reviews.words(fileid)), category)
                 for category in reviews.categories()
                 for fileid in reviews.fileids(category)]
    random.shuffle(documents)

    # Hàm trích xuất đặc trưng
    all_words = nltk.FreqDist(w.lower() for w in reviews.words())
    word_features = list(all_words.keys())[:2000]

    def document_features(document):
        document_words = set(document)
        features = {}
        for word in word_features:
            features[f'contains({word})'] = (word in document_words)
        return features

    # Tạo featuresets
    featuresets = [(document_features(d), c) for (d, c) in documents]

    # Chia train/test
    split = int(len(featuresets) * 0.8)
    train_set, test_set = featuresets[:split], featuresets[split:]

    # Huấn luyện
    classifier = NaiveBayesClassifier.train(train_set)
    acc = nltk_accuracy(classifier, test_set)
    print(f"Độ chính xác Naive Bayes classifier: {acc:.2%}")
    print(f"\nCác đặc trưng quan trọng nhất:")
    classifier.show_most_informative_features(10)

except Exception as e:
    print(f"Lỗi: {e}")
    print("(Có thể do thiếu dữ liệu movie_reviews)")

```
